```python
from langchain.schema import HumanMessage
from langchain_ollama import ChatOllama


server_url = "http://localhost:11434"

ollama_model = ChatOllama(model="llama3.1", server=server_url)

# ارسال پیام ساده به مدل
response = ollama_model.invoke([HumanMessage(content="سلام! آیا می‌توانی خودت را معرفی کنی؟")])
response2 = ollama_model.invoke([HumanMessage(content="Hi, would you introduce yourself?")])
# نمایش پاسخ
print("Model Response:", response , response2)
```
